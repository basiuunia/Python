```python
import folium
import pandas as pd
```

Szybkie wgranie mapy podkłądowej

```python
#Współrzedne w ukłdzie WGS84, czyli geograficzne
mapa = folium.Map(location=[52.40, 16.9],
               zoom_start=10 # stopień przybliżenia skala od 1 do 18
              )
mapa.add_child(folium.LatLngPopup())   
mapa
```

Wgranie punktów ze współrzędnymi

```python
zrm=pd.read_csv('./dane_geo/zrm.csv', sep =';')
zrm.head(10)
```

```python
mapa = folium.Map(location=[52.40, 16.9],
               zoom_start=10 
              )
for index, row in zrm.iterrows():
    folium.Marker([row['LAT'], row['LONG']], 
                  popup=row['adres'],
                  tooltip=row['nazwa']
                 ).add_to(mapa)
    
 
mapa
```
